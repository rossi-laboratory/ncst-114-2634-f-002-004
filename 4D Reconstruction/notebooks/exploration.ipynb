```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('.').resolve().parent))

import numpy as np
import matplotlib.pyplot as plt
from utils.io_utils import load_npz_tracks
```

```python
# Load a tracked sequence and inspect basic stats.
data = load_npz_tracks('../data/point_tracks/sample.npz')
tracks = data['tracks']  # (T, N, 3)
print('shape:', tracks.shape, ' valid frac:', float(np.isfinite(tracks).all(-1).mean()))
```

```python
# Plot per-frame point Z distribution.
plt.figure(figsize=(8, 3))
plt.imshow(tracks[..., 2].T, aspect='auto', cmap='viridis')
plt.xlabel('frame')
plt.ylabel('point id')
plt.colorbar(label='depth (z)')
plt.title('Per-point depth across time')
plt.tight_layout()
plt.show()
```
